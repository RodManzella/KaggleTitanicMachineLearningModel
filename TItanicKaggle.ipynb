# Creating dataframes

```python
gender_sub_path = 'datasets/gender_submission.csv'
test_path = 'datasets/test.csv'
train_path = 'datasets/train.csv'

gender_submission_df = pd.read_csv(gender_sub_path)
test_df = pd.read_csv(test_path)
train_df = pd.read_csv(train_path)
```

# Creating a dataframe with only people who did not survive

```python
did_not_survive = train_df.loc[train_df['Survived'] == 0] 
did_not_survive
```

# Creating a dataframe with only people who did survive


```python
did_survive = train_df.loc[train_df['Survived'] == 1] 
did_survive
```

# Analysis based on age

## Histogram of people who survived based on age

```python
plt.hist(did_survive['Age'], bins=50, edgecolor='black')  # 'bins' define o número de barras no histograma
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age histogram')
plt.grid(True)

# Exibir o gráfico
plt.show()
```

## Histogram of people who died based on age

```python
plt.hist(did_not_survive['Age'], bins=50, edgecolor='black')  # 'bins' define o número de barras no histograma
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age histogram')
plt.grid(True)

# Exibir o gráfico
plt.show()
```
